```python
import streamlit as st
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
from PIL import Image
```

```python
# Memuat model
model = load_model('../best_model(85_74).h5')
```

```python
# Fungsi untuk memproses gambar dan melakukan prediksi
def predict_image(img, model):
    img = img.resize((224, 224))  # Sesuaikan dengan ukuran input model Anda
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Normalisasi jika diperlukan oleh model
    prediction = model.predict(img_array)
    return prediction
```

```python

# Fungsi untuk halaman utama
def main_page():
    st.markdown(
        """
        <style>
        .header {
            background-color: #FFD700;
            padding: 20px;
            border-radius: 10px;
        }
        .header h1 {
            font-family: 'Arial Black', sans-serif;
            font-size: 50px;
            text-align: center;
        }
        .header p {
            text-align: center;
        }
        .upload-box {
            display: flex;
            justify-content: center;
            align-items: center;
            height: 300px;
            border: 2px dashed #CCCCCC;
            border-radius: 10px;
        }
        </style>
        """,
        unsafe_allow_html=True
    )

    # Header section with background color
    st.markdown(
        """
        <div class="header">
            <h1>SiBetta</h1>
            <p>Teman Setia Pecinta Cupang!</p>
            <p>Aplikasi web inovatif yang menggunakan teknologi AI untuk membantu Anda mengidentifikasi jenis ikan cupang dengan cepat dan mudah. Aplikasi ini dirancang untuk pedagang ikan, pembeli ikan cupang, dan pecinta ikan cupang di seluruh Indonesia.</p>
        </div>
        """,
        unsafe_allow_html=True
    )

    st.write("")
    st.write("## Mulai Sekarang!!!")

    # Upload box for images
    uploaded_file = st.file_uploader("Masukkan Gambar", type=["jpg", "jpeg", "png"])
    if uploaded_file is not None:
        img = Image.open(uploaded_file)
        st.image(img, caption='Gambar yang diunggah.', use_column_width=True)
        
        st.write("Mengolah gambar...")
        prediction = predict_image(img, model)
        
        # Simpan hasil prediksi ke sesi state
        st.session_state['prediction'] = prediction
        st.session_state['uploaded_img'] = img
        st.session_state['page'] = 'result'
```

```python
# Fungsi untuk halaman hasil prediksi
def result_page():
    if 'prediction' in st.session_state and 'uploaded_img' in st.session_state:
        img = st.session_state['uploaded_img']
        prediction = st.session_state['prediction']
        
        st.markdown(
            """
            <style>
            .header {
                background-color: #FFD700;
                padding: 20px;
                border-radius: 10px;
            }
            .header h1 {
                font-family: 'Arial Black', sans-serif;
                font-size: 50px;
                text-align: center;
            }
            .header p {
                text-align: center;
            }
            .content {
                text-align: center;
            }
            </style>
            """,
            unsafe_allow_html=True
        )

        # Header section with background color
        st.markdown(
            """
            <div class="header">
                <h1>Crown Tail</h1>
            </div>
            """,
            unsafe_allow_html=True
        )

        st.image(img, caption='Gambar yang diunggah.', use_column_width=True)
        st.write("Rp10.000 - Rp500.000 per ekor.")
        st.write("")
        st.write("Karakteristik Ikan Cupang Crown Tail:")
        st.write("- Sirip: Sirip ekor cupang Crown Tail panjang dan mengalir, dengan tepi yang bergerigi atau berenda. Sirip ini menyerupai mahkota saat dikembangkan.")
        st.write("- Warna: Cupang Crown Tail memiliki berbagai macam warna, termasuk merah, biru, hijau, ungu, dan hitam. Warna-warnanya seringkali cerah dan berkilauan.")
        st.write("- Ukuran: Cupang Crown Tail adalah ikan cupang yang relatif kecil, dengan panjang tubuh sekitar 5 cm.")
        st.write("- Temperamen: Cupang Crown Tail adalah ikan yang agresif, terutama terhadap jantan lain. Mereka tidak boleh dipelihara bersama ikan cupang jantan lainnya.")
        st.write("- Habitat: Cupang Crown Tail berasal dari Asia Tenggara dan ditemukan di air tawar yang tenang, seperti sawah dan kolam.")
        st.write("- Perawatan: Cupang Crown Tail membutuhkan akuarium yang kecil namun terawat dengan baik. Mereka membutuhkan air yang hangat dan bersih, serta banyak tempat untuk bersembunyi.")
        st.write("")
        st.write("Secara umum, harga ikan cupang Crown Tail di Indonesia berkisar antara Rp10.000 hingga Rp500.000 per ekor.")
        st.write("")
        st.write("Berikut adalah beberapa contoh harga ikan cupang Crown Tail di Indonesia:")
        st.write("- Rp10.000 - Rp20.000: Ikan cupang Crown Tail muda atau betina dengan kualitas rendah.")
        st.write("- Rp20.000 - Rp50.000: Ikan cupang Crown Tail dewasa atau jantan dengan kualitas rendah.")
        st.write("- Rp50.000 - Rp100.000: Ikan cupang Crown Tail dengan kualitas sedang.")
        st.write("- Rp100.000 - Rp250.000: Ikan cupang Crown Tail dengan kualitas tinggi.")
        st.write("- Rp250.000 - Rp500.000: Ikan cupang Crown Tail dengan kualitas super tinggi, seperti yang memiliki garis keturunan juara atau yang telah memenangkan kontes.")

# Tentukan halaman mana yang akan ditampilkan
if 'page' not in st.session_state:
    st.session_state['page'] = 'main'

if st.session_state['page'] == 'main':
    main_page()
elif st.session_state['page'] == 'result':
    result_page()
```
